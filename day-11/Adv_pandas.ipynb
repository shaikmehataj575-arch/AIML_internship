IMPORTING PANDAS

```python
import pandas as pd
```

READING DATASET : HEALTHCARE DATASET

```python
df = pd.read_csv("healthcare_dataset.csv.zip")
print(df.head())
```

1. GroupBy() : 
groupby() is a Pandas function used to group rows that have the same value in one or more columns and then perform calculations like sum, average, count, etc.

Grouping Patients based on Medical Condition: Here patients are grouped according to their medical condition and we count how many patients belong to each condition. This helps hospitals understand which diseases are most common.

```python

condition_count = df.groupby("Medical Condition")["Name"].count()
print(condition_count)
```

Average Billing Amount by Insurance Provider : This groups patients based on their insurance provider and calculates the average billing amount for each provider. It helps analyze which insurance companies handle higher medical expenses and supports financial decision-making.

```python
avg_bill = df.groupby("Insurance Provider")["Billing Amount"].mean()
print(avg_bill)
```

Applying Aggregation with GroupBy

```python
grouping = df.groupby(["Gender", "Medical Condition"])
print(grouping.first())
```

Applying Aggregation with GroupBy

```python
grouped_data = df.groupby(["Gender", "Medical Condition"]).agg(
    Patient_Count=("Name", "count"),
    Avg_Billing=("Billing Amount", "mean")
)

print(grouped_data)
```

Filtering groups

```python
filtered_df = df.groupby("Medical Condition").filter(
    lambda x: x["Billing Amount"].mean() > 3000
)
print(filtered_df.head())
```

Transformation Methods

```python
df["Condition_Avg_Bill"] = df.groupby("Medical Condition")["Billing Amount"].transform("mean")
print(df[["Medical Condition", "Billing Amount", "Condition_Avg_Bill"]].head())
```

2. MERGE() : merge() is a Pandas function used to combine two DataFrames based on a common column.

CREATING ANOTHER DATASET

```python
df1 = pd.DataFrame({
    "Hospital": ["Sons and Miller","Reed LLC","Newton-White"],
    "Rating": [4.5, 4.2, 4.8]
})
```

Inner Merge : An inner merge retains only the rows where the merge keys exist in both DataFrames

```python
merged_df = pd.merge(
    df,
    df1,
    on="Hospital",
    how="inner"
)

print(merged_df)
```

Left merge : A left merge retains every row from the left DataFrame, along with only the matching rows from the right DataFrame

```python
merged_df = pd.merge(
    df,
    df1,
    on="Hospital",
    how="left"
)

print(merged_df)
```

Right merge : A right merge retains every row from the right DataFrame, along with only the matching rows from the left DataFrame

```python
merged_df = pd.merge(
    df,
    df1,
    on="Hospital",
    how="right"
)

print(merged_df)
```

outer merge : An outer merge (or full outer join) retains all rows from both the left and right DataFrames

```python
merged_df = pd.merge(
    df,
    df1,
    on="Hospital",
    how="outer"
)

print(merged_df)
```

Left on and Right on merge functions : When two DataFrames contain related information but the matching columns have different names, left_on and right_on tell Pandas which columns should be used for matching. This allows us to combine data without renaming columns first.

```python
df1 = pd.DataFrame({
    "Hospital_Name": ["Sons and Miller","Reed LLC","Newton-White"],
    "Rating": [4.5, 4.2, 4.8]
})
```

```python
merged_df = pd.merge(
    df,
    df1,
    left_on="Hospital",
    right_on="Hospital_Name",
    how="inner"
)

print(merged_df)
```

3. PIVOT() : pivot() (or pivot_table()) is a Pandas function used to reshape and summarize data by converting rows into columns.

Average Billing Amount by Gender and Admission Type

```python
pivot_table = df.pivot_table(
    index="Gender",
    columns="Admission Type",
    values="Billing Amount",
    aggfunc="mean"
)

print(pivot_table)
```

Multi-Level Pivot Table

```python
pivot_table = df.pivot_table(
    index="Gender",
    columns="Admission Type",
    values=["Billing Amount","Age"]
)

print(pivot_table)
```
