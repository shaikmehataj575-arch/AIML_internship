1.Bar plots

```python
quality_counts = df['quality'].value_counts().sort_index()

plt.figure(figsize=(12,7))
plt.bar(
    quality_counts.index,
    quality_counts.values,
    color='mediumseagreen',
    edgecolor='black',
    linewidth=2)
plt.title('Wine Quality Distribution', fontsize=18, fontweight='bold')
plt.xlabel('Quality Rating')
plt.ylabel('Number of Wines')
plt.show()
```

2.Histogram with KDE

```python
sns.set_style("whitegrid")
numerical_columns = df.select_dtypes(include=['int64','float64']).columns
plt.figure(figsize=(18,25))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(df[feature],kde=True,color='pink',edgecolor='black')
plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")
plt.show()
```

3.Swarm Plot 

```python
plt.figure(figsize=(10, 8))
sns.swarmplot(x="quality", y="alcohol", data=df, palette='viridis')
plt.title('Swarm Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

 4.Pair Plot

```python
sns.set_palette("Pastel1")
plt.figure(figsize=(10, 6))
sns.pairplot(df)
plt.suptitle('Pair Plot for DataFrame')
plt.show()
```

5.Violin Plot 

```python
plt.figure(figsize=(14,8))
sns.violinplot(x='quality',y='alcohol',data=df,palette='coolwarm',linewidth=2)
plt.title('Alcohol vs Quality (Violin Plot)',)
plt.xlabel('Quality', fontsize=14)
plt.ylabel('Alcohol', fontsize=14)
plt.show()
```
