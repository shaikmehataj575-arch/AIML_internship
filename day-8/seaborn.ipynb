1.Line Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Name': ['AKRITI', 'SAMREEN', 'RAYAN', 'RAGHAV'],
    'Age': [21,25,23,29]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8,5))
sns.lineplot(x=df.index,y='Age',data=df,marker='*',markersize= 20,color='purple',linewidth=3)
plt.title("Age Trend Analysis")
plt.xlabel("Student Index")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```

2.Scatter Plot

```python
df = pd.DataFrame(data)
plt.figure(figsize=(8,5))
sns.scatterplot(x=df.index,y='Age',data=df,s=200,color='red',alpha=0.7)
plt.title("Age Distribution")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()
```

3.Box Plot

```python
plt.figure(figsize=(7,5))
sns.boxplot(y='Age',data=df,color='orange',width=0.4)
plt.title("Age Box Plot")
plt.show()
```

4.Violin Plot

```python

plt.figure(figsize=(8,5))
sns.violinplot(y='Age',data=df,color='skyblue',linewidth=2)
plt.title("Age Violin Plot")
plt.show()
```

5.Swarm Plot

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x=df.index,y='Age',data=df,size=10,color='green')
plt.title("Swarm Plot of Ages")
plt.show()
```

6.Bar Plot

```python
plt.figure(figsize=(9,5))
sns.barplot(x='Name',y='Age',data=df,palette='Set2')
plt.title("Student Age Comparison")
plt.xlabel("Students")
plt.ylabel("Age")
plt.show()
```

7.Count Plot

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Name',data=df,palette='coolwarm')
plt.title("Frequency of Student Names")
plt.xlabel("Names")
plt.ylabel("Count")
plt.show()
```

8.KDE Plot

```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = load_iris()
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)
plt.figure(figsize=(9,5))
sns.kdeplot(
    data=df,
    x='sepal length (cm)',
    fill=True,
    color='magenta',
    linewidth=2
)
plt.title("Sepal Length Density Curve")
plt.show()
```

Customize Seaborn Plots 

1. Adding Titles and Axis Labels

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

2. Built-in Styles and Grids in Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid")

sns.boxplot(x='species', y='petal_length', data=sns.load_dataset('iris'),color = 'yellow')
plt.title('Petal Length Distribution by Species')
plt.show()
```

3. Customizing Color Palettes

```python
custom_colors = ["#D0042A", '#33FFBD', "#D51BEA"]
sns.violinplot(x='species', y='petal_length', data=iris,palette=custom_colors,)
plt.title('Custom Colored Petal Length Distribution')
plt.show()
```

1.Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
sns.set_style("darkgrid")
sns.pairplot(iris,hue="species",palette="Set2",diag_kind="kde",height=2.5)
plt.suptitle("Modified Iris Pair Plot", y=1.02)
plt.show()
```

2.Joint Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
sns.regplot(x='total_bill', y='tip', data=tips, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.show()
```

3.Grid Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.set_style("ticks")
g = sns.FacetGrid(
    tips,
    col="time",
    row="sex",
    height=4,
    aspect=1.2
)

g.map(
    sns.scatterplot,
    "total_bill",
    "tip",
    color="darkgreen"
)

g.fig.suptitle(
    "Total Bill vs Tip by Time and Gender",
    y=1.05,
    fontsize=16
)

plt.show()
```
