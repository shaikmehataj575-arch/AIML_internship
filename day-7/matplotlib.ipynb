Data Visualization using Matplotlib in Python

1. Line Chart

```python
import matplotlib.pyplot as plt
import pandas as pd
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel(x)
plt.ylabel(y)
plt.grid()
plt.show()
```

2. Bar Chart

```python
data = {"salary":[25000,40000,35000,30000,50000,20000,25000,50000,60000,10000],
        "dept":["HR","Finance","IT","HR","IT","HR","IT","Finance","IT","IT"]
        }
df = pd.DataFrame(data)
print(df)
count = df["dept"].value_counts()
print(count)
plt.bar(count.index,count,color = 'grey')
plt.title("Department Frequency")
plt.xlabel("department")
plt.ylabel("salary")
plt.show()
```

 3.Histogram

```python
plt.hist(df["salary"],bins = 5,color = "skyblue")
plt.title("Histogram")
plt.xlabel("salary")
plt.ylabel("No of Employees")
plt.show()
```

4. Scatter Plot

```python
df["age"] = [23,25,29,35,45,26,44,30,40,25]
plt.scatter(df["salary"],df["age"])
plt.title("Scatter Plot")
plt.xlabel("salary")
plt.ylabel("age")
print(df)
plt.grid()
plt.show()
```

5. Pie Chart

```python
tot_sal_by_dept = df.groupby("dept")["salary"].sum()
print(tot_sal_by_dept)
plt.pie(tot_sal_by_dept,labels=tot_sal_by_dept.index,autopct = "%1.0f",explode=[0,0.1,0])
plt.title("Pie Chart")
plt.show()
```

6. Box Plot

```python
data = [[10,12,15,19,21,22,29,11],[2,4,5,6,3,8,10,7],[12,19,20,22,28,23,25,29]]
plt.boxplot(data)
plt.xlabel("Groups")
plt.ylabel("Values")
plt.title("Box Plot")
plt.show()
```

7. Heatmap

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
data = np.random.rand(10, 10)

plt.imshow(data)

plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap')
plt.show()
```

Customize Matplotlib Visualizations

1. Customizing Line Chart

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y,color = "green",linewidth = 3,marker = "*",linestyle = ":",markersize = 20)
plt.title("Line Graph")
plt.xlabel(x)
plt.ylabel(y)
plt.grid()
plt.show()
```

2. Customizing Bar Chart

```python
plt.bar(count.index,count,color = ["violet","pink","red"],width=0.5,linewidth = 2,edgecolor = "black")
plt.title("Department Frequency")
plt.xlabel("department")
plt.ylabel("salary")
plt.show()
```

3. Customizing Histogram Plot

```python
plt.hist(df["salary"],bins = 5,color = "yellow",edgecolor = "red",linestyle = "dashed",alpha = 1)
plt.title("Histogram")
plt.xlabel("salary")
plt.ylabel("No of Employees")
plt.show()
```

4. Customizing Scatter Plot

```python
df["age"] = [23,25,29,35,45,26,44,30,40,25]
plt.scatter(df["salary"],df["age"], marker='D', alpha=0.5)
plt.title("Scatter Plot")
plt.xlabel("salary")
plt.ylabel("age")
plt.show()
```

5. Customizing Pie Chart

```python
plt.pie(tot_sal_by_dept,labels=tot_sal_by_dept.index,autopct = "%1.0f",explode=[0.4,0.1,0],shadow = True)
plt.title("Pie Chart")
plt.show()
```

Matplotlib’s Core Components: Figures and Axes

```python
fig = plt.figure(figsize=(6, 4), facecolor='teal')
ax = fig.add_axes([0.1, 1.0, 1.0, 1.0])
x = [1, 2, 3, 4,5]
y = [100,250,150,300,250]
ax.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

2. Axes Class

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

x = [10,20,30,40]
y = [23,25,35,55]
z = [20,39,49,60]

fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0.1,0.1,0.8,0.8]) 

ax.plot(x,y)
ax.plot(y,z)
ax.plot(x,z)

ax.set_title("Linear Graph")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.legend(labels=('line 1', 'line 2','line 3'))
plt.show()
```

Advanced Techniques for Visualizing Subplots

1. Using add_axes()

```python
from matplotlib.figure import Figure
x = [100,129,300,499,900]
y = [200,300,500,900,1000]
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_axes([0.1,0.1,0.35,0.8])
ax2 = fig.add_axes([0.55,0.1,0.35,0.8])
ax1.plot(x,y)
ax2.plot(y,x)
plt.show()
```

2. Using subplot()

```python
x = [100,129,300,499,900]
y = [200,300,500,900,1000]
plt.figure()
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(y,x)
plt.show()
```

3. Using subplot2grid()


```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]
axes1 = plt.subplot2grid (
(8, 1), (0, 0), rowspan = 2, colspan = 1)
axes2 = plt.subplot2grid (
(7, 1), (2, 0), rowspan = 2, colspan = 1)

axes1.plot(x, y)
axes2.plot(y, x)
plt.show()
```

Saving Plots Using savefig()

```python
year = ['2010', '2002', '2004', '2006', '2008']
production = [25, 15, 35, 30, 10]

plt.bar(year, production)

plt.savefig("output.jpg")
plt.savefig("output1", facecolor='y', bbox_inches="tight",pad_inches=0.3, transparent=True)
```
